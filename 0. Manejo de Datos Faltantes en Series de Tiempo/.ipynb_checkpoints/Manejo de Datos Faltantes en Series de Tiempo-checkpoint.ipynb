```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def limpiar_outliers_iqr(df_mensual, factor_iqr=1.5, hacer_grafico=True):
    """
    Detecta outliers con IQR sobre Quantity y los reemplaza por valores interpolados en el tiempo.
    - factor_iqr: 1.5 (estándar), puedes usar 2.0 o 3.0 si quieres ser menos estricto.
    - hacer_grafico: True para ver gráfico antes/después.
    Devuelve:
        df_out: DataFrame con columnas ['YearMonth','Quantity','Quantity_clean','is_outlier']
        info: dict con Q1, Q3, IQR, lower, upper y meses outliers
    """

    # Copia ordenada por fecha
    df = df_mensual.copy()
    df = df.sort_values('YearMonth').reset_index(drop=True)

    # IQR global sobre toda la serie
    q = df['Quantity']
    Q1, Q3 = q.quantile(0.25), q.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - factor_iqr * IQR
    upper = Q3 + factor_iqr * IQR

    # Marcar outliers
    df['is_outlier'] = (q < lower) | (q > upper)

    # Reemplazar outliers por NaN y luego interpolar en el tiempo
    df['Quantity_clean'] = df['Quantity'].where(~df['is_outlier'], np.nan)

    # Interpolación temporal (requiere índice datetime)
    df_idx = df.set_index('YearMonth')
    try:
        # Si la frecuencia es mensual, esto hará una interpolación por tiempo
        df_idx['Quantity_clean'] = df_idx['Quantity_clean'].interpolate(method='time')
    except Exception:
        # Fallback a lineal si el método 'time' no es aplicable
        df_idx['Quantity_clean'] = df_idx['Quantity_clean'].interpolate(method='linear')

    # Bordes: si quedó NaN al inicio/fin, rellenar con vecino más cercano
    df_idx['Quantity_clean'] = df_idx['Quantity_clean'].bfill().ffill()

    # Volver a columna
    df_out = df_idx.reset_index()

    # (Opcional) Gráfico comparación
    if hacer_grafico:
        plt.figure(figsize=(12,6))
        plt.plot(df_out['YearMonth'], df_out['Quantity'], marker='o', label='Original')
        plt.plot(df_out['YearMonth'], df_out['Quantity_clean'], marker='s', label='Limpia')
        # Marcar outliers originales
        xo = df_out.loc[df_out['is_outlier'], 'YearMonth']
        yo = df_out.loc[df_out['is_outlier'], 'Quantity']
        plt.scatter(xo, yo, marker='x', s=80, label='Outliers detectados')
        plt.title('Serie original vs. serie limpiada (IQR + interpolación)')
        plt.xlabel('Mes'); plt.ylabel('Quantity')
        plt.grid(True); plt.legend(); plt.xticks(rotation=45)
        plt.show()

    info = {
        'Q1': float(Q1), 'Q3': float(Q3), 'IQR': float(IQR),
        'lower': float(lower), 'upper': float(upper),
        'meses_outliers': df_out.loc[df_out['is_outlier'], 'YearMonth'].dt.strftime('%Y-%m').tolist()
    }
    return df_out, info

# ===== Uso =====
df_limpio, info_out = limpiar_outliers_iqr(df_mensual, factor_iqr=1.5, hacer_grafico=True)

print("Bandas IQR:", {k: round(v,2) for k,v in info_out.items() if k in ['Q1','Q3','IQR','lower','upper']})
print("Meses marcados como outliers:", info_out['meses_outliers'])

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

# ========= 1. Preparar la serie =========
# Asegúrate que YearMonth sea datetime y sea el índice
serie = df_mensual.set_index('YearMonth')['Quantity']
serie = serie.asfreq('MS')  # frecuencia mensual

# ========= 2. Ajustar un modelo SARIMA =========
# (p,d,q)(P,D,Q,12) → 12 porque es frecuencia anual
modelo = sm.tsa.statespace.SARIMAX(serie,
                                   order=(1,1,1),
                                   seasonal_order=(1,1,1,12),
                                   enforce_stationarity=False,
                                   enforce_invertibility=False)

resultado = modelo.fit(disp=False)

# ========= 3. Predicciones y residuales =========
pred = resultado.predict(start=serie.index[0], end=serie.index[-1])
residuales = serie - pred

# ========= 4. Umbral para outliers =========
umbral = 2 * np.std(residuales)  # puedes usar 3 si quieres menos estrictos
outliers = residuales[abs(residuales) > umbral]

print("Meses detectados como outliers:")
print(outliers)

# ========= 5. Visualización =========
plt.figure(figsize=(12,6))

# Serie original
plt.plot(serie.index, serie.values, label='Real', marker='o')

# Predicción del modelo
plt.plot(pred.index, pred.values, label='Predicho (SARIMA)', linestyle='--')

# Outliers (rojo)
plt.scatter(outliers.index, serie.loc[outliers.index],
            color='red', marker='x', s=100, label='Outliers detectados')

plt.title("Detección de outliers con SARIMA")
plt.xlabel("Mes"); plt.ylabel("Quantity")
plt.legend(); plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from prophet import Prophet

# ========= 1. Preparar datos para Prophet =========
# Prophet exige columnas 'ds' (fecha) y 'y' (valor)
df_prophet = df_mensual[['YearMonth','Quantity']].rename(columns={'YearMonth':'ds','Quantity':'y'})

# ========= 2. Entrenar modelo Prophet =========
modelo = Prophet(yearly_seasonality=True, weekly_seasonality=False, daily_seasonality=False)
modelo.fit(df_prophet)

# ========= 3. Predicciones =========
forecast = modelo.predict(df_prophet[['ds']])
df_prophet['yhat'] = forecast['yhat']

# ========= 4. Residuales =========
df_prophet['residual'] = df_prophet['y'] - df_prophet['yhat']

# Umbral de outliers (2 desviaciones estándar)
umbral = 2 * np.std(df_prophet['residual'])
df_prophet['is_outlier'] = abs(df_prophet['residual']) > umbral

outliers = df_prophet[df_prophet['is_outlier']]

print("Meses detectados como outliers con Prophet:")
print(outliers[['ds','y','yhat','residual']])

# ========= 5. Visualización =========
plt.figure(figsize=(12,6))

# Serie real
plt.plot(df_prophet['ds'], df_prophet['y'], marker='o', label='Real')

# Predicción Prophet
plt.plot(df_prophet['ds'], df_prophet['yhat'], linestyle='--', label='Predicho (Prophet)')

# Outliers
plt.scatter(outliers['ds'], outliers['y'], color='red', marker='x', s=100, label='Outliers detectados')

plt.title("Detección de outliers con Prophet")
plt.xlabel("Mes"); plt.ylabel("Quantity")
plt.legend(); plt.grid(True)
plt.show()
```

```python
import pandas as pd
from typing import Optional, Tuple, Union

def preparar_series_mensuales(
    csv_path: str,
    max_date: Union[str, pd.Timestamp],
    *,
    start_date: Optional[Union[str, pd.Timestamp]] = None,
    date_col: str = "DocDate",
    sku_col: str = "ItemCode",
    qty_col: str = "Quantity",
    freq: str = "M",            # "M" (fin de mes) o "MS" (inicio de mes)
    fill_missing: float = 0,    # 0 si mes sin venta = 0; usa None para dejar NaN
    return_wide: bool = False,  # True para devolver también formato ancho
) -> Union[pd.DataFrame, Tuple[pd.DataFrame, pd.DataFrame]]:
    """
    Lee un CSV de ventas, consolida por fecha+SKU y devuelve una serie mensual
    alineada para todos los SKUs hasta 'max_date'.

    Retorna (por defecto) un DataFrame LONG con MultiIndex (Fecha, SKU) y columna 'Qty'.
    Si return_wide=True, retorna (long, wide) donde wide tiene Fecha como índice y SKUs como columnas.
    """
    # 1) Carga
    df = pd.read_csv(csv_path)

    # 2) Tipos y limpieza mínima
    df = df[[date_col, sku_col, qty_col]].copy()
    df[date_col] = pd.to_datetime(df[date_col], errors="coerce")
    df[sku_col]  = df[sku_col].astype(str)
    df[qty_col]  = pd.to_numeric(df[qty_col], errors="coerce").fillna(0)

    # Eliminar filas sin fecha o sin SKU
    df = df.dropna(subset=[date_col, sku_col])
    if df.empty:
        raise ValueError("No hay filas válidas tras la limpieza (fechas/SKU faltantes).")

    # 3) Consolidar por fecha+SKU (por si hay múltiples transacciones en el mismo día)
    df = df.groupby([date_col, sku_col], as_index=False)[qty_col].sum()

    # 4) Índice jerárquico y orden
    df = df.set_index([date_col, sku_col]).sort_index()

    # 5) Determinar rango temporal
    #    - start_date: inicio del rango (si no se pasa, usa la mínima fecha de datos)
    #    - max_date  : fin del rango (tope de análisis)
    if start_date is None:
        first = df.index.get_level_values(date_col).min()
    else:
        first = pd.to_datetime(start_date)

    last = pd.to_datetime(max_date)

    # Normalizar al mes correcto según freq
    if freq.upper() == "M":   # fin de mes
        first = first.replace(day=1)  # para que el primer 'M' sea el fin de ese primer mes
    elif freq.upper() == "MS":  # inicio de mes
        first = first.replace(day=1)
        last  = last.replace(day=1)

    full_range = pd.date_range(first, last, freq=freq)
    if len(full_range) == 0:
        raise ValueError("El rango temporal quedó vacío. Revisa start_date, max_date y freq.")

    # 6) Resample mensual por SKU + reindex común
    mensual = (
        df[qty_col]
          .groupby(level=sku_col)
          .resample(freq, level=date_col)
          .sum()
          .reindex(full_range, level=date_col)  # alinea todos los SKUs al mismo rango
          .to_frame("Qty")
    )

    if fill_missing is not None:
        mensual["Qty"] = mensual["Qty"].fillna(fill_missing)

    # (DocDate, ItemCode) como orden del índice
    mensual = mensual.swaplevel(0, 1).sort_index()

    if return_wide:
        wide = mensual["Qty"].unstack(sku_col)
        if fill_missing is not None:
            wide = wide.fillna(fill_missing)
        return mensual, wide

    return mensual
```

```python
def preparar_series_mensuales_robusto(
    csv_path, max_date, *,
    start_date=None,
    date_col="DocDate", sku_col="ItemCode", qty_col="Quantity",
    freq="M", fill_missing=0, return_wide=True
):
    import pandas as pd

    df = pd.read_csv(csv_path, usecols=[date_col, sku_col, qty_col])
    df[date_col] = pd.to_datetime(df[date_col], errors="coerce")
    # Fuerza SKU como texto (preserva ceros a la izquierda)
    df[sku_col]  = df[sku_col].astype(str)
    df[qty_col]  = pd.to_numeric(df[qty_col], errors="coerce").fillna(0)
    df = df.dropna(subset=[date_col, sku_col])

    # Consolida por día y SKU
    df = df.groupby([date_col, sku_col], as_index=False)[qty_col].sum()
    df = df.set_index([date_col, sku_col]).sort_index()

    # Rango mensual fijo
    first = df.index.get_level_values(date_col).min() if start_date is None else pd.to_datetime(start_date)
    last  = pd.to_datetime(max_date)
    # Normaliza a mes
    first = first.replace(day=1)
    if freq.upper() == "MS":
        last = last.replace(day=1)
    full_range = pd.date_range(first, last, freq=freq)

    # --- Paso clave robusto: ancho -> reindex -> largo ---
    wide = (
        df[qty_col]
        .groupby(level=sku_col)
        .resample(freq, level=date_col)
        .sum()
        .unstack(sku_col)              # filas=meses, columnas=SKU
        .reindex(full_range)           # añade TODOS los meses hasta 'last'
    )
    if fill_missing is not None:
        wide = wide.fillna(fill_missing)

    mensual_long = (
        wide
        .stack(sku_col)
        .to_frame("Qty")
        .rename_axis(index=[date_col, sku_col])
        .sort_index()
    )

    return (mensual_long, wide) if return_wide else mensual_long
```

```python
import pandas as pd

# Parámetros
dataframe = 'SALES_IMPORTACEROS_19082025.csv'
fecha_desde = '2023-01-01'
fecha_hasta = '2024-12-31'

# 1. Carga de dataframe
df_original = pd.read_csv(dataframe)

# 2. Copia del df original para trabajar
df = df_original.copy()

# 3. Solo extraigo las columnas para proyección
df = df[['DocDate', 'ItemCode', 'Quantity']]

# 4. Transformo columna fecha
df['DocDate'] = pd.to_datetime(df['DocDate'])

# 5. Agrupo cantidades por día y item
df = df.groupby(['ItemCode', 'DocDate'])['Quantity'].sum().reset_index()

# 6. Filtro el rango global de fechas para análisis
df = df[(df['DocDate'] >= fecha_desde) & (df['DocDate'] <= fecha_hasta)]

# 7. Función para completar fechas **desde la primera venta real** hasta fecha_hasta
def completar_calendario(grupo):
    # Fecha de inicio = mayor entre la primera venta real y fecha_desde global
    fecha_inicio = max(grupo['DocDate'].min(), pd.to_datetime(fecha_desde))
    calendario = pd.date_range(start=fecha_inicio, end=fecha_hasta, freq='D')
    
    # Reindexar el grupo con el calendario dinámico
    grupo = grupo.set_index('DocDate').reindex(calendario, fill_value=0)
    
    # Restaurar columnas
    grupo = grupo.rename_axis('DocDate').reset_index()
    grupo['ItemCode'] = grupo['ItemCode'].iloc[0]  # mantener el código del SKU
    return grupo

# 8. Aplicamos la función por cada producto
df = df.groupby('ItemCode', group_keys=False).apply(completar_calendario)

# 9. Reordenamos columnas y establecemos índice por fecha
df = df[['DocDate', 'ItemCode', 'Quantity']].set_index('DocDate')

# 10. Visualizamos resultado
print(df.head(20))
```

```python
import pandas as pd

def preparar_series_tiempo(
    file_path,
    fecha_desde,
    fecha_hasta,
    col_fecha='DocDate',
    col_sku='ItemCode',
    col_qty='Quantity'
):
    """
    Prepara un DataFrame de series de tiempo por SKU, completando fechas faltantes
    dinámicamente desde la primera venta de cada SKU hasta fecha_hasta.
    
    Devuelve un DataFrame en formato largo con índice temporal.
    """
    
    # 1. Carga del CSV
    df = pd.read_csv(file_path)
    
    # 2. Selección de columnas
    df = df[[col_fecha, col_sku, col_qty]]
    
    # 3. Conversión a datetime
    df[col_fecha] = pd.to_datetime(df[col_fecha])
    
    # 4. Agrupamos cantidades por día y SKU
    df = df.groupby([col_sku, col_fecha])[col_qty].sum().reset_index()
    
    # 5. Filtro rango global
    df = df[(df[col_fecha] >= fecha_desde) & (df[col_fecha] <= fecha_hasta)]
    
    # 6. Función interna para completar calendario dinámico por SKU
    def completar_calendario(grupo):
        fecha_inicio = max(grupo[col_fecha].min(), pd.to_datetime(fecha_desde))
        calendario = pd.date_range(start=fecha_inicio, end=fecha_hasta, freq='D')
        grupo = grupo.set_index(col_fecha).reindex(calendario, fill_value=0)
        grupo = grupo.rename_axis(col_fecha).reset_index()
        grupo[col_sku] = grupo[col_sku].iloc[0]
        return grupo
    
    # 7. Aplicamos por SKU
    df_final = df.groupby(col_sku, group_keys=False).apply(completar_calendario)
    
    # 8. Reordenamos columnas y ponemos índice temporal
    df_final = df_final[[col_fecha, col_sku, col_qty]].set_index(col_fecha)
    
    return df_final
```

### Situación 3: marcas de tiempo faltantes "ocultas"

Esta situación es similar a la anterior. Sin embargo, en principio ninguna marca de tiempo aparece etiquetada como *NaT*. Son fechas que no existen en el df

Vemos que hay un "santo de fechas" entre 2008-04-03 y 2008-04-05. Es decir que la Serie de Tiempo tiene una frecuencia diaria pero esta frecuencia no se mantiene para 2008-04-04.

En este caso Pandas lee la Serie de Tiempo de manera correcta y **no aparecerán marcas faltantes etiquetadas con *NaT***.

### Detección usando el método "difference"

En este caso la idea es:

1. Crear un rango ideal de fechas
2. Usar "difference" para encontrar las diferencias entre este rango y el índice original del DataFrame

```python
# 1. Crear un rango completo de fechas
rango_fechas = pd.date_range(start=df.index.min(), end=df.index.max(), freq='D')
rango_fechas
```

```python
# 2. Usar "difference" para encontrar las diferencias entre este rango y el índice original del *DataFrame*
rango_fechas.difference(df.index)
```
